15.Find cube and cube root of a number using reusable function

18.Calculate Compound interest

```python
p = float(input())
r = float(input())
t = float(input())

amount = p*((1+r/100)**t)
CI = amount - p
print("Compound Interest:",CI)
```

Another method using reusable function

```python
def compound_interest(p,r,t):
  amount = p*((1+r/100)**t)
  CI = amount -p
  return CI

a = float(input())
b = float(input())
c = float(input())

print("Compound Interest:",round(compound_interest(a,b,c),3))
```
