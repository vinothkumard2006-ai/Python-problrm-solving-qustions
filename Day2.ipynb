3.Find grade from marks

```python
def student_marks(mark):
   if mark >= 90:
      return "A"
   elif mark >=80:
      return "B"
   elif mark >= 70:
      return "C"
   elif mark >= 60:
      return "D"
   else :
      return "C"

mark = int(input())

print("Student Grade:",student_marks(mark))
```

4.Check divisibility by 5 and 11

```python
def divisible(num):
   if num % 5 ==0 and num % 11 ==0:
       return "Divisible"
   else:
       return "Not Divisible"

num = (int(input()))
print("Divisible by 5 and 11 :",divisible(num))
```

6.Check alphabet or not

```python
def check_alphabet(ch):
   if 'a' <= ch <= 'z' or 'A' <= ch <= 'z':
       return "Alphabet"
   else:
       return "Not alphabet"

char = input()
print("Check Alphabet:",check_alphabet(char))
```

10.Electricity bill calculation

```python
def bill(units):
   if units <= 100:
     bill = units*1.5
     return bill
   elif units <= 200:
     bill = 100*1.5 + (units - 100)*2.5
     return bill
   else :
     bill = 100*1.5 + 100*2.5 + (units - 200)*4
     return bill

units = int(input())
print("Electricity Bill:",bill(units))
```

12.Check whether year is a century

```python
def century(year):
   if year % 100 ==0:
      return "Century"
   else:
      return "Not Century"

year = int(input())
print(f"{year} is a {century(year)}")
```

13.Palindrome

```python
def palindrome(text):
   cleaned = text.lower().replace(" ", "")
   if cleaned == cleaned[::-1]:
      return "Palindrome"
   else:
      return "Not Palindrome"

text = input()
print(palindrome(text))
```
