1. Even or odd number

```python
num = int(input())

if num % 2 ==0:
  print("Even number")
else:
  print("Odd number")
```

2. Check Prime Number

```python
num = int(input())

if num <= 1:
  print("Not Prime")
else:
  for i in range(2,num):
     if num % i== 0:
       print("Not Prime")
       break
  else:
    print("Prime")
```

3. Factorial

```python
n = int(input())

fact =1
for i in range(1,n+1):
   fact*=i

print(fact)
```

4. Fibonacci Series

```python
n = int(input())
a,b =0,1
for i in range(n):
  print(a,end=" ")
  a, b = b, a+b
```

5. Armstrong number

```python
num = int(input())

temp = num
s=0
while temp>0:
  digits = temp%10
  s+=digits**3
  temp//=10
if s==num:
  print("Amstrong number")
else:
  print("Not a Armstrong number")
```

6. Palindrome number

```python
n = input()
if n == n[::-1]:
  print("Palindrome number")
else:
  print("Not a palindrome")
```

7. Reverse string

```python
n = input()
print(n[::-1])
```

8. Count vowels

```python
s = input().lower()

count =0
for ch in s:
  if ch in "aeiou":
    count+=1
print(count)
```

9. Second largest element in list

```python
nums = list(map(int,input().split()))

nums = list(set(nums))
nums.sort()
print(nums[-2])
```

10. Remove duplicate from list

```python
nums = list(map(int,input().split()))

print("Remove duplicates")
print(list(set(nums)))
```
