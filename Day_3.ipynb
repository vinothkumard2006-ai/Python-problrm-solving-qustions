2.Check triangle validity

3.Palindrome number

```python
num = input()

if num == num[::-1]:
   print("Palindrome number")
else:
   print("Not a Palindrome number")
```

4.Amstrong number

```python
num = 153
temp = num

ams = 0

while temp>0:
    digits = temp%10
    ams += digits**3
    temp//=10

if ams == num:
   print("Its Amstrong number")
else:
   print("Not a Amstrong number")
```
